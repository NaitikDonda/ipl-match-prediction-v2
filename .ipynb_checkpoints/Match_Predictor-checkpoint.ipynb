# ============================================================
# PART 4: FEATURE ENGINEERING
# ============================================================

```python
match_df["team1_is_toss_winner"] = (match_df["toss_winner"] == match_df["team1"]).astype(int)
match_df["team2_is_toss_winner"] = (match_df["toss_winner"] == match_df["team2"]).astype(int)
match_df["toss_decision_bat"] = (match_df["toss_decision"] == "bat").astype(int)

print("Basic toss features created")
```

```python
match_df["match_month"] = match_df["date"].dt.month
match_df["match_day_of_week"] = match_df["date"].dt.dayofweek

match_df["season"] = match_df["season"].str[:4]  # take only first year
match_df["season"] = match_df["season"].fillna(match_df["date"].dt.year).astype(int)

print("Temporal features created")
```

```python
match_df_sorted = match_df.sort_values("date").reset_index(drop=True)

h2h_wins_team1 = []
h2h_total = []
h2h_win_rate = []

for i in range(len(match_df_sorted)):
    row = match_df_sorted.iloc[i]
    t1, t2 = row["team1"], row["team2"]

    past = match_df_sorted.iloc[:i]
    past = past[
        ((past["team1"] == t1) & (past["team2"] == t2)) |
        ((past["team1"] == t2) & (past["team2"] == t1))
    ]

    total = len(past)
    wins = (past["match_won_by"] == t1).sum()

    h2h_total.append(total)
    h2h_wins_team1.append(wins)
    h2h_win_rate.append(wins / total if total > 0 else 0.5)

match_df_sorted["h2h_wins_team1"] = h2h_wins_team1
match_df_sorted["h2h_total_matches"] = h2h_total
match_df_sorted["h2h_win_rate_team1"] = h2h_win_rate

print("Head-to-head features done")
```

```python
N_FORM = 5

def get_form(team, date):
    past = match_df_sorted[
        (match_df_sorted["date"] < date) &
        ((match_df_sorted["team1"] == team) | (match_df_sorted["team2"] == team))
    ].tail(N_FORM)

    if len(past) == 0:
        return 0.5

    return (past["match_won_by"] == team).sum() / len(past)


form_team1 = []
form_team2 = []

for i in range(len(match_df_sorted)):
    row = match_df_sorted.iloc[i]

    form_team1.append(get_form(row["team1"], row["date"]))
    form_team2.append(get_form(row["team2"], row["date"]))


match_df_sorted["form_team1_last5"] = form_team1
match_df_sorted["form_team2_last5"] = form_team2
match_df_sorted["form_diff"] = match_df_sorted["form_team1_last5"] - match_df_sorted["form_team2_last5"]

print("Recent form features done")
```

```python
venue_win_team1 = []
venue_win_team2 = []

for i in range(len(match_df_sorted)):
    row = match_df_sorted.iloc[i]
    t1, t2 = row["team1"], row["team2"]
    venue = row["venue"]

    past = match_df_sorted.iloc[:i]
    past = past[past["venue"] == venue]

    # team1
    t1_matches = past[(past["team1"] == t1) | (past["team2"] == t1)]
    if len(t1_matches) == 0:
        venue_win_team1.append(0.5)
    else:
        venue_win_team1.append((t1_matches["match_won_by"] == t1).mean())

    # team2
    t2_matches = past[(past["team1"] == t2) | (past["team2"] == t2)]
    if len(t2_matches) == 0:
        venue_win_team2.append(0.5)
    else:
        venue_win_team2.append((t2_matches["match_won_by"] == t2).mean())


match_df_sorted["venue_wr_team1"] = venue_win_team1
match_df_sorted["venue_wr_team2"] = venue_win_team2

print("Venue win rate features done")
print("Shape:", match_df_sorted.shape)
```

```python
feature_cols_preview = [
    "match_id", "team1", "team2", "season",
    "toss_winner", "toss_decision", "venue", "city",
    "team1_is_toss_winner", "toss_decision_bat",
    "h2h_win_rate_team1", "h2h_total_matches",
    "form_team1_last5", "form_team2_last5", "form_diff",
    "venue_wr_team1", "venue_wr_team2",
    "match_month", "match_day_of_week",
    "match_won_by"
]
feature_cols_preview = [c for c in feature_cols_preview if c in match_df_sorted.columns]

print("Feature sample:")
print(match_df_sorted[feature_cols_preview].head(10))

match_df = match_df_sorted.copy()
```

```python
def predict_ipl_winner(team1, team2, toss_winner, toss_decision, venue, city):

    # auto season
    season = match_df["season"].max()
    match_history = match_df[match_df["season"] <= season]

    team1_is_toss = int(toss_winner == team1)
    toss_bat = int(toss_decision == "bat")

    # head-to-head
    h2h = match_history[
        ((match_history["team1"] == team1) & (match_history["team2"] == team2)) |
        ((match_history["team1"] == team2) & (match_history["team2"] == team1))
    ]
    h2h_wr = (h2h["match_won_by"] == team1).mean() if len(h2h) > 0 else 0.5

    # form
    def get_form(team):
        m = match_history[
            (match_history["team1"] == team) | (match_history["team2"] == team)
        ].tail(5)
        return (m["match_won_by"] == team).mean() if len(m) > 0 else 0.5

    form1 = get_form(team1)
    form2 = get_form(team2)

    # venue
    venue_data = match_history[match_history["venue"] == venue]

    def venue_wr(team):
        m = venue_data[
            (venue_data["team1"] == team) | (venue_data["team2"] == team)
        ]
        return (m["match_won_by"] == team).mean() if len(m) > 0 else 0.5

    v1 = venue_wr(team1)
    v2 = venue_wr(team2)

    # input
    data = pd.DataFrame([{
        "team1": team1,
        "team2": team2,
        "venue": venue,
        "city": city,
        "toss_decision": toss_decision,
        "team1_is_toss_winner": team1_is_toss,
        "toss_decision_bat": toss_bat,
        "h2h_win_rate_team1": h2h_wr,
        "h2h_total_matches": len(h2h),
        "form_team1_last5": form1,
        "form_team2_last5": form2,
        "form_diff": form1 - form2,
        "venue_wr_team1": v1,
        "venue_wr_team2": v2,
        "match_month": 4,
        "match_day_of_week": 5,
        "season": season
    }])

    X = preprocessor.transform(data)
    pred = model.predict(X)[0]
    prob = model.predict_proba(X)[0]

    return {
        "Winner": team1 if pred == 1 else team2,
        "Team1 %": round(prob[1]*100, 2),
        "Team2 %": round(prob[0]*100, 2)
    }
```

```python
print("Sample Predictions\n" + "=" * 40)

sample_fixtures = [
    {
        "team1": "Mumbai Indians", "team2": "Chennai Super Kings",
        "toss_winner": "Mumbai Indians", "toss_decision": "bat",
        "venue": "Wankhede Stadium", "city": "Mumbai"
    },
    {
        "team1": "Royal Challengers Bengaluru", "team2": "Kolkata Knight Riders",
        "toss_winner": "Kolkata Knight Riders", "toss_decision": "field",
        "venue": "Eden Gardens", "city": "Kolkata"
    },
    {
        "team1": "Rajasthan Royals", "team2": "Sunrisers Hyderabad",
        "toss_winner": "Rajasthan Royals", "toss_decision": "bat",
        "venue": "Sawai Mansingh Stadium", "city": "Jaipur"
    },
    {
        "team1": "Delhi Capitals", "team2": "Punjab Kings",
        "toss_winner": "Punjab Kings", "toss_decision": "field",
        "venue": "Arun Jaitley Stadium", "city": "Delhi"
    },
]

for i, f in enumerate(sample_fixtures, 1):
    result = predict_ipl_winner(**f)

    print(f"\nMatch {i}: {f['team1']} vs {f['team2']}")
    print(f"Venue : {f['venue']}, {f['city']}")
    print(f"Toss  : {f['toss_winner']} → {f['toss_decision']}")

    print("Predicted Winner:", result["Winner"])
    print(f"{f['team1']} Win %: {result['Team1 %']:.2f}")
    print(f"{f['team2']} Win %: {result['Team2 %']:.2f}")

    print("-" * 40)
    print("Model Accuracy:", round(accuracy_score(y_test, model.predict(X_test_proc))*100, 2), "%")
```

```python
def predict_live_win_prob(runs_scored, balls_bowled, wickets_fallen, runs_target):

    balls_left = 120 - balls_bowled
    runs_left = runs_target - runs_scored
    wickets_left = 10 - wickets_fallen

    # simple features
    X = np.array([[runs_left, balls_left, wickets_left]])
    
    prob = live_model.predict_proba(X)[0]

    return {
        "Chasing %": round(prob[1]*100, 2),
        "Defending %": round(prob[0]*100, 2),
        "Status": f"{runs_left} runs needed in {balls_left} balls"
    }
```

```python
print("Live Predictions")

scenarios = [
    (142, 90, 2, 162),
    (137, 108, 6, 162),
    (102, 90, 7, 162),
    (82, 60, 4, 162)
]

for s in scenarios:
    result = predict_live_win_prob(*s)

    print("\nNeeded:", result["Status"])
    print("Chasing Team %:", result["Chasing %"])
    print("Defending Team %:", result["Defending %"])
```
